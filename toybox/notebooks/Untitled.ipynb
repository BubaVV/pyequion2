```python
def dict_print(*dicts):
    #Assume dicts have same keys
    keys = dicts[0].keys()
    if len(keys) == 0:
        m = 0
    else:
        m = max(len(key) for key in keys)
    for key in keys:
        n = len(key)
        print_str = ' '*(m-n) + '{0}:'.format(key) + ''.join([' {0:.4e}'.format(d[key]) for d in dicts])
        print(print_str)
```

```python
#Case 1: basic equilibria
eqsys = EquilibriumSystem(['C','Ca','Na','Cl'], from_elements=True, activity_model="PITZER")
elements_balance = {'Ca':0.028, 'Cl':0.056, 'Na':0.075, 'C':0.065}
TK = 300
solution,_ = eqsys.solve_equilibrium_elements_balance(TK, elements_balance, tol=1e-12)
dict_print(solution.molals,solution.activities)
print()
dict_print(solution.elements_molals)
print()
dict_print(solution.solid_molals,solution.saturation_indexes)
```

```python
#Case 2: Equilibria with solid precipitation
eqsys = EquilibriumSystem(['C','Ca','Na','Cl'], from_elements=True, activity_model="PITZER")
elements_balance = {'Ca':0.028, 'Cl':0.056, 'Na':0.075, 'C':0.065}
TK = 300
solution,_ = eqsys.solve_equilibrium_elements_balance_solids(TK, elements_balance, tol=1e-12)
dict_print(solution.molals,solution.activities)
print()
dict_print(solution.elements_molals)
print()
dict_print(solution.solid_molals,solution.saturation_indexes)
```

```python
#Case 3: Equilibria fixing species
eqsys = EquilibriumSystem(['C','Ca','Na','Cl'], from_elements=True, activity_model="PITZER")
molal_balance = {'Ca++':0.028, 'Cl-':0.056, 'Na+':0.075, 'HCO3-':0.065}
TK = 298.15
solution,_ = eqsys.solve_equilibrium_mixed_balance(TK, molal_balance=molal_balance, tol=1e-12)
dict_print(solution.molals,solution.activities)
print()
dict_print(solution.elements_molals)
print()
dict_print(solution.solid_molals,solution.saturation_indexes)
```
